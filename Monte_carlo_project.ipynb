```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
initial_investment = 10000
days = 252
mean_return = 0.0005
volatility = 0.02
num_simulations = 100
final_values = []
```

```python
for simulation in range(num_simulations):
    portfolio = initial_investment

    for day in range(days):
        daily_return = np.random.normal(mean_return, volatility)
        portfolio = portfolio * (1 + daily_return)

    final_values.append(portfolio)
```

```python
print("Average final value:", np.mean(final_values))
print("Min:", np.min(final_values))
print("Max:", np.max(final_values))
```

```python
plt.hist(final_values, bins = 20)
plt.title("Monte Carlo Simulation")
plt.xlabel("Final Portfolio Value")
plt.ylabel("Frequency")
plt.show()
```
